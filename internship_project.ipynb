# Left and right of numpy array & convert it into image

```python
# Left shift of an numpy array
t = np.array([[ 92, 171, 212],
  [ 92, 171, 212],
  [ 92, 171, 212]])
u = np.array([[199, 201, 213],
  [201, 204, 213],
  [202, 205, 212]]) 
v = np.left_shift(t,u)

v
```

```python
# displaying the image 

img.res = plt.imshow(v)
```

```python
# Right shift of an image

m = np.array( [[ 92, 171, 212],
  [ 92, 171, 212],
  [ 92, 171, 212]])
  
n = np.array([[ 93, 172, 213],
  [ 93, 172, 213],
  [ 93, 172, 213]])

z = np.right_shift(m,n)
```

```python
img_res = plt.imshow(z)
```
